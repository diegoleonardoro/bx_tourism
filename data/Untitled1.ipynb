```python
sum (df.head()["NUMBER OF SALES"])
```

```python
bx_neighborhoods = df ["NEIGHBORHOOD"].unique()
```

```python
dict_ = {}

for n in bx_neighborhoods:
    x = df[df["NEIGHBORHOOD"] == n]
    total_num_sales = sum (x["NUMBER OF SALES"])
    neighborhood = x["NEIGHBORHOOD"].iloc[0]
    dict_ [neighborhood] = total_num_sales
```
